# 概要

- numpyおよびpyplotモジュールを使用しました

```python
import numpy
import matplotlib.pyplot as plt
```

# 1. ベルヌーイ分布

### 1.1 ベルヌーイ分布に従ったサンプルを返す関数

```python
def Bern(p, n):
    return np.random.choice([0,1], size=n, p=[(100-p)/100,p/100]) 
```

```python
print(Bern(p=50,n=80))
```

### 1.2 ベルヌーイ分布をプロットする関数

```python
def plot_Bern(data):
    unique, unique_counts = np.unique(data,return_counts=True)
    unique_counts = unique_counts/len(data)*100
    plt.bar(unique, unique_counts)
    plt.xticks(unique)
    plt.ylim([0,100])
    plt.xlabel("Outcome")
    plt.ylabel("Frequency(%)")
    plt.title("Bernoulli Distribution")
    plt.show()
```

```python
sample = Bern(p=70,n=100)
plot_Bern(sample)
```

# 2. カテゴリカル分布

### 2.1 カテゴリカル分布に従ったサンプルを返す関数

```python
def Cat(p_vals,n):
    p_list = np.arange(len(p_vals))
    p_vals = np.array(p_vals)/100
    return np.random.choice(a=p_list,size=n,p=p_vals)
```

```python
Cat(p_vals=[10,80,10],n=80)
```

### 2.2 カテゴリカル分布をプロットする関数

```python
def plot_Cat(data):
    unique, unique_counts = np.unique(data,return_counts=True)
    unique_counts = unique_counts/len(data)*100
    plt.bar(unique, unique_counts)
    plt.xticks(unique)
    plt.xlabel("Outcome")
    plt.ylabel("Frequency(%)")
    plt.title("Categorical Distribution")
    plt.show()
```

```python
sample = Cat(p_vals=[10,80,10],n=1000)
plot_Cat(sample) 
```

# 3. 正規分布

### 3.1 正規分布に従ったサンプルを返す関数

```python
def Normal(m,s,n):
    return s * np.random.randn(n) + m
```

```python
Normal(m=100,s=1,n=20)
```

### 3.2 連続分布をプロットする関数

```python
def plot_Cont(sample):
    plt.hist(sample,bins=50,density=True)
    plt.xlabel("Value")
    plt.ylabel("Density(%)")
    plt.title("Continuous Distribution"),
    plt.show()
```

```python
sample=Normal(1,1,1000000)
plot_Cont(sample)
```
